```python
def match_positive(pattern, target):
    if len(pattern) != len(target):
        return False
    for m, c in zip(pattern, target):
        if m != '_' and m != c:
            return False
    return True
```

```python
match_positive('he__o', 'hello')
```

```python
def match(pattern, target, excluded=None):
    if not excluded:
        excluded = ''
    if len(pattern) != len(target):
        return False
    for m, c in zip(pattern, target):
        if m == '_' and c not in excluded:
            # true
            pass
        elif m != '_' and m == c:
            # true
            pass
        else:
            return False
    return True
```

```python
match('__l_o', 'hello', 'def')
```
